```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
## Sigmoid Function

def Sigmoid(x):
  a=[]
  for i in x:
    a.append(1/(1+math.exp(-i)))
  return a
```

```python
## Hyperbolic Tangent

def Tanh(x, derivate=False):
  if(derivate==True):
    return (1-x**2)
  else:
    return np.tanh(x)
```

```python
## ReLU

def ReLU(x):
  a = []
  for i in x:
    if(i<0):
      a.append(0)
    else:
      a.append(i)
  return a
```

```python
## Leakly ReLU

def LeakyReLU(x):
  a = []
  for i in x:
    if(i<0):
      a.append(i/10)
    else:
      a.append(i)
  return a
```

```python
## Swish

def swish(x):
    return x * Sigmoid(x)
```

```python
##

x = np.arange(-3, 3., 0.1)

sig = Sigmoid(x)
tanh = Tanh(x)
relu = ReLU(x)
leakyrelu = LeakyReLU(x)
swish = swish(x)


line1 = plt.plot(x, sig, label='Sigmoid')
line2 = plt.plot(x, tanh, label='Tanh')
line3 = plt.plot(x, relu, label='ReLU')
line4 = plt.plot(x, leakyrelu, label='LeakyReLU')
line5 = plt.plot(x, swish, label='Swish')

plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
