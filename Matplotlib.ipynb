# https://matplotlib.org

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yasListesi = [10, 20, 30, 30,40, 40, 50, 60, 70, 75]
kiloListesi = [20, 60, 80, 85, 86, 87, 70, 90, 95, 90]

numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
numpyYasListesi

```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") ## g green anlamına gelmekte yeşil çizgi ile gösteriyor.
plt.title("Yas kilo orani")
plt.xlabel("Yas")
plt.ylabel("Kilo")
```

```python
numpyDizisi1 = np.linspace(0,10,20)

numpyDizisi1

numpyDizisi2 = numpyDizisi1 ** 3

numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2,"g--") ##-- koyarak kesikli çizgi haline getirdik // g*- ,
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")

plt.subplot(1, 2, 2)  # 1 sıra olacak, 2 kolon olacak, şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi2, numpyDizisi1, "g--")
```

```python
benimFigure = plt.figure() ##figure boş bir kanvastır
figureAxes = benimFigure.add_axes([0.4, 0.4, 0.9, 0.9]) ##x ekseninde hareket , y ekseninde hareket, width, height (bu şekilde x ve y eksenleri ekliyoruz)
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1, 0.1, 0.7, 0.7])
eksen2 = figure2.add_axes([0.1, 0.3, 0.3, 0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout() ##Bu oluşan grafiklerin birbiri içerisine girmesini engelliyor
```

```python
yeniFigur = plt.figure() ##PARANTEZİN İÇİNE FARKLI DEĞERLER VEREBİLİRİZ ÜSTÜNE GELİP BAK
yeniEksen = yeniFigur.add_axes([0.1, 0.1, 1, 1])
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 2, label="numpyDizisi ** 2") ##alpha=0.3 gibi değerler vererek saydamlığı değiştirebiliriz linewidth=0.5 çizgi kalınlığı
                                                                          ##linestyle="--" çizgi stili
                                                                          ##marker = "+" gibi markerler (kesişme noktaları koyabiliriz) markercolor,markerstyle var
yeniEksen.plot(numpyDizisi1, numpyDizisi2 ** 3, label="numpyDizisi ** 3")
yeniEksen.legend() ##bu yukarıdaki çizgi bilgilerini göstermeye yarıyor. parantezinin içine loc=0 gibi bir sayı verirsek lokasyonu değişir
```

```python
yeniFigur.savefig("BenimFigure.png")##dpi resim kalitesi
```
