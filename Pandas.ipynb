```python

yeniBirVeri = {
    "Karakter Sinifi": ["South Park", "South Park", "Simpson", "Simpson"],
    "Karakter Ismi": ["Cartman", "Kenny", "Homer", "Bart"],
    "Karakter Yas": [9, 10, 50, 20]
}

karakterDF = pd.DataFrame(yeniBirVeri)
karakterDF.pivot_table(values="Karakter Yas", index=["Karakter Sinifi", "Karakter Ismi"])  ##burada parantezin içine aggfunc = np.sum verirsek aynı olan kişilerde değerleri toplar. farklı fonksiyonlarda verebiliriz
```


Bu kod örneğinde, çizgi film karakterlerinin sınıfları, isimleri ve yaşları ile ilgili verileri içeren bir pandas DataFrame oluşturuluyor. Veri seti, yeniBirVeri adlı bir sözlükle tanımlanıyor ve bu sözlük üç anahtara sahip:

"Karakter Sinifi": Karakterlerin ait olduğu çizgi filmi temsil eder (örneğin, "South Park", "Simpson").
"Karakter Ismi": Çizgi film karakterlerinin adlarını içerir (örneğin, "Cartman", "Kenny", "Homer", "Bart").
"Karakter Yas": Karakterlerin yaşlarını içerir (örneğin, 9, 10, 50, 20).
Bu verilerle bir DataFrame (karakterDF) oluşturulduktan sonra, pivot_table fonksiyonu kullanılarak karakterlerin yaş ortalamaları (veya başka istatistiksel değerler) hesaplanıyor. pivot_table çağrısında:

values="Karakter Yas": Hesaplamalar için kullanılacak değerlerin sütunu belirtiliyor (burada yaşlar).
index=["Karakter Sinifi", "Karakter Ismi"]: Pivot tablonun indeksleri olarak hangi sütunların kullanılacağını belirtir. Bu durumda, çizgi film sınıfı ve karakter ismi iki seviyeli bir indeks olarak kullanılıyor.
Sonuç olarak, bu pivot tablo her bir çizgi film sınıfı ve karakter kombinasyonu için yaş ortalamasını (bu örnekte her karakter için doğrudan yaşları, çünkü tek tek kayıtlar var) gösterir. İleri seviyede kullanımlarda aggfunc parametresi ile istenilen hesaplama fonksiyonu belirtilebilir, örneğin np.mean, np.sum gibi.
